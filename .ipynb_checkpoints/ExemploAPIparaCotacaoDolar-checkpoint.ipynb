# API Exemplo
## Quermos pegar a cotação do dólar de forma automática e atualizada, como faremos?

Usamos o awesome api: https://docs.awesomeapi.com.br/

API do Twilio comentada: https://www.twilio.com/docs/libraries/python

FONTE: https://www.youtube.com/watch?v=eel1OVIdfUw&t=264s

```python
#Biblioteca para fazer requisição
#Caso não tenha, para instalar é >pip install requests
import requests

#também é necessário importar a biblioteca JSON
import json

#pegar as informações das dotações
cotacoes = requests.get("https://economia.awesomeapi.com.br/last/USD-BRL,EUR-BRL,BTC-BRL")
#formata o resultado no formato json
cotacoes = cotacoes.json()
#alimenta com a cotacao do dolar, se colocar apenas o USDBRL, ele lista tudo
#mas se colocar o bid, ele busca a cotação do momento, pois o bid é alimentado com essa informação
cotacao_dolar = cotacoes['USDBRL']['bid']

print("Todas as Cotacoes: ",cotacoes)
print()
print("Cotacao do dolar está em $%s em %s" %(cotacao_dolar,cotacoes['USDBRL']['create_date']))
```
